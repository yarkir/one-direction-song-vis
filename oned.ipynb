```python
import pandas as pd
df=pd.read_csv('One_Direction_All_Songs_wr.csv')
df.head(10)
```

```python
import matplotlib.pyplot as plt #plt - конвенциональный алиас
import seaborn as sns
```

```python
albums=df.groupby('Album(s)', as_index=False).count()
albums.head(10)
```

```python
plt.figure(figsize=(12, 8))
plt.grid(visible=True, color='gray', linestyle='--', linewidth=0.5)
plt.barh(albums['Album(s)'], albums['Song'])
plt.title('Количество песен в каждом альбоме')
plt.xlabel('Название альбома')
plt.ylabel('Количество песен')
#количество песен в каждом альбоме
```

```python
god= df.groupby('Year', as_index=False).count()
god
```

```python
plt.figure(figsize=(12, 8))
plt.grid()
sns.lineplot(x="Year", y="Song",
             data=god)
plt.title('Количество песен за каждый год')
plt.xlabel('Год')
plt.ylabel('Количество песен')

```

```python
value_counts = df['Writer(s)'].value_counts().reset_index()
value_counts.columns = ['Writer(s)', 'Count']

plt.figure(figsize=(60, 14))
sns.barplot(x='Writer(s)', y='Count', data=value_counts)


plt.title('Количество произведений по авторам')
plt.xticks(rotation=45)
#какие авторы писали песни и сколько песен написал каждый автор
```
